```python
import tarfile

# Spcifiez le chemin vers le fichier .tgz  dcompresser
tgz_file = "demo/data/bofip_stock_live_20250129.tgz"

# Spcifiez le rpertoire o extraire le contenu
extract_path = "demo/data/uncompressed/"

# Ouvrir le fichier .tgz en mode lecture compresse (gzip)
with tarfile.open(tgz_file, "r:gz") as tar:
    # (Optionnel) Afficher la liste des fichiers contenus dans l'archive
    tar.list()
    # Extraire tous les fichiers dans le rpertoire spcifi
    tar.extractall(path=extract_path)

print("Dcompression termine !")
```
