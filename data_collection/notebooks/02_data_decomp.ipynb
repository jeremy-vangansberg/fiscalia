```python
import tarfile

# Spécifiez le chemin vers le fichier .tgz à décompresser
tgz_file = "demo/data/bofip_stock_live_20250129.tgz"

# Spécifiez le répertoire où extraire le contenu
extract_path = "demo/data/uncompressed/"

# Ouvrir le fichier .tgz en mode lecture compressée (gzip)
with tarfile.open(tgz_file, "r:gz") as tar:
    # (Optionnel) Afficher la liste des fichiers contenus dans l'archive
    tar.list()
    # Extraire tous les fichiers dans le répertoire spécifié
    tar.extractall(path=extract_path)

print("Décompression terminée !")
```
