```python
import requests
from bs4 import BeautifulSoup
import logging
from typing import Optional, List
from pathlib import Path
import os

session = requests.Session()
headers = {
            'User-Agent': 'Mozilla/5.0 (compatible; BOFiPBot/1.0)'
        }
url = "https://data.economie.gouv.fr/explore/dataset/bofip-impots/information/"

response = session.get(url, headers=headers)
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')
# Trouve tous les liens contenant 'bofip_documentation' dans leur href
links = soup.find_all('a', href=lambda href: 'bofip_documentation' in href)

print(links)
```
